```python
def find_records_with_difference(grade, predicted_grade, model_answer, student_answer, question, n):
    result_records = []

    for q, sa, ma, g, pg in zip(question, student_answer, model_answer, grade, predicted_grade):
        if abs(g - pg) == n:
            result_records.append((q, sa, ma, g, pg))

    return result_records
```

```python
# Example usage:
grades = [5, 4, 2, 3, 5]
predicted_grades = [3, 3, 2, 1, 4]
model_answers = ["A", "B", "A", "C", "B"]
student_answers = ["A", "C", "A", "C", "B"]
questions = ["Q1", "Q2", "Q3", "Q4", "Q5"]
requested_difference = 2

result = find_records_with_difference(grades, predicted_grades, model_answers, student_answers, questions, requested_difference)

print(result)
```
